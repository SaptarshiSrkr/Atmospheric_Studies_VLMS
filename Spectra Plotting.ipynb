```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lhs72 = pd.read_csv('Data/OBSERVED/Processed/norm_RVcorr_LHS72.txt',names=['wave','flux'],delim_whitespace=True)
lhs73 = pd.read_csv('Data/OBSERVED/Processed/norm_RVcorr_LHS73.txt',names=['wave','flux'],delim_whitespace=True)
```

```python
synthlhs72 = pd.read_csv('Parameter_Estimation/Results/LHS72/2-August/conv1.34_norm_lte39.8-4.70-1.4a+X.X.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
synthlhs73 = pd.read_csv('Parameter_Estimation/Results/LHS73/7-August/conv1.34_norm_lte39.6-4.82-1.4a+X.X.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
```

```python
plt.figure(figsize=(15,10))
plt.plot(lhs72.wave,lhs72.flux,'k',lw=1,label='Observed')
plt.plot(synthlhs72.wave,synthlhs72.flux,'r',lw=1,label='MCMC Best Fit')
plt.xlim(6500,8200)
plt.legend()
plt.savefig('LHS72_blue_arm.png',dpi=500,facecolor='white')
```

```python
plt.figure(figsize=(15,10))
plt.plot(lhs73.wave,lhs73.flux,'k',lw=1,label='Observed')
plt.plot(synthlhs73.wave,synthlhs73.flux,'r',lw=1,label='MCMC Best Fit')
plt.xlim(6500,8200)
plt.legend()
plt.savefig('LHS73_blue_arm.png',dpi=500,facecolor='white')
```

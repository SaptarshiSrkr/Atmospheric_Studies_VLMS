```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
synth1 = pd.read_csv('Data/SYNTHETIC/Convolved_Abun/conv1.34_norm_Fe_722_lte040-4.70-1.3a+0.4.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
synth2 = pd.read_csv('Data/SYNTHETIC/Convolved_Abun/conv1.34_norm_Fe_752_lte040-4.70-1.3a+0.4.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
synth3 = pd.read_csv('Data/SYNTHETIC/Convolved_Abun/conv1.34_norm_Fe_792_lte040-4.70-1.3a+0.4.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)

synth4 = pd.read_csv('Data/SYNTHETIC/Convolved_Abun/conv1.34_norm_Ca_600_lte040-4.70-1.3a+0.4.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
synth5 = pd.read_csv('Data/SYNTHETIC/Convolved_Abun/conv1.34_norm_Ca_634_lte040-4.70-1.3a+0.4.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
```

```python
lhs72 = pd.read_csv('Data/OBSERVED/Processed/norm_RVcorr_LHS72.txt',names=['wave','flux'],delim_whitespace=True)
lhs73 = pd.read_csv('Data/OBSERVED/Processed/norm_RVcorr_LHS73.txt',names=['wave','flux'],delim_whitespace=True)
```

```python
synthlhs72 = pd.read_csv('Parameter_Estimation/Results/LHS72/2-August/conv1.34_norm_lte39.8-4.70-1.4a+X.X.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
synthlhs73 = pd.read_csv('Parameter_Estimation/Results/LHS73/7-August/conv1.34_norm_lte39.6-4.82-1.4a+X.X.BT-Settl.CIFIST2011_2017.spec.7.txt',names=['wave','flux'],delim_whitespace=True)
```

```python
plt.figure(figsize=(15,10))
plt.plot(synth1.wave,synth1.flux,lw=1,label='Fe_722')
plt.plot(synth2.wave,synth2.flux,lw=1,label='Fe_752')
plt.plot(synth3.wave,synth3.flux,lw=1,label='Fe_792')

plt.plot(synth4.wave,synth4.flux,lw=1,label='Ca_600')
plt.plot(synth5.wave,synth5.flux,lw=1,label='Ca_634')

plt.xlim(6501,6502)
plt.ylim(0.2,0.5)
plt.legend()
plt.savefig('LHS72_blue_arm.png',dpi=500,facecolor='white')
```

```python
import emcee
import matplotlib.pyplot as plt
```

```python
reader = emcee.backends.HDFBackend('logfile_norm_RVcorr_LHS72.txt.h5')
```

```python
samples = reader.get_chain(flat=True)
```

```python
import corner
figure = corner.corner(samples,labels=['Teff', 'Log g', '[M/H]'],quantiles=[0.16, 0.5, 0.84], show_titles=True, title_kwargs={"fontsize": 12}, range=[(3800,4200),(4.0,5.5),(-2.5,0)])
#plt.savefig(f'Corner_norm_RVcorr_LHS72.txt.png',dpi=500,facecolor='white')
```

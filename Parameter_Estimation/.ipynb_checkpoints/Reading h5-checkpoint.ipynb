```python
import emcee
import numpy as np
import matplotlib.pyplot as plt
```

```python
reader = emcee.backends.HDFBackend('logfile_norm_RVcorr_LHS72.txt.h5')
```

```python
samples = reader.get_chain(flat=True)
```

```python
import corner
figure = corner.corner(samples,labels=['Teff', 'Log g', '[M/H]'],quantiles=[0.16, 0.5, 0.84], show_titles=True, title_kwargs={"fontsize": 12}, range=[(3800,4200),(4.0,5.5),(-2.5,0)])
#plt.savefig(f'Corner_norm_RVcorr_LHS72.txt.png',dpi=500,facecolor='white')
```

```python
fig, ax = plt.subplots(3, sharex=True,figsize=(10,8))
for i in range(3):
    ax[i].plot(reader.get_chain()[:, :, i], '-k', alpha=0.5)

ax[0].set_ylabel('Teff',fontsize=15)
ax[1].set_ylabel('log g',fontsize=15)
ax[2].set_ylabel('[M/H]',fontsize=15)
#fig.savefig(f'Chains_{observed_file}_final.png',dpi=500)
```
